```python
#Average drinks per week by binge drinking status
BRFSS2023_df.groupby('_rfbing6')['_drnkwk2'].mean()
```

```python
#Flu shot status by binge drinking
BRFSS2023_df.groupby('_rfbing6')['_flshot7'].value_counts(normalize=True)

```

```python
#Drunk driving risk by binge drinking
BRFSS2023_df.groupby('_rfbing6')['_drnkdrv'].mean()
```

```python
plt.hist(BRFSS2023_df['_drnkwk2'].dropna(), bins=30)
plt.xlabel("Drinks per Week")
plt.ylabel("Count")
plt.title("Distribution of Weekly Alcohol Consumption")
plt.show()
```

```python
BRFSS2023_df.groupby('_rfbing6')['_flshot7'].mean().plot(kind='bar')
plt.xlabel("Binge Drinking (0=No, 1=Yes)")
plt.ylabel("Flu Shot Rate")
plt.title("Flu Shot Rate by Binge Drinking Status")
plt.show()
```

```python
plt.scatter(BRFSS2023_df['_drnkwk2'], BRFSS2023_df['_drnkdrv'], alpha=0.3)
plt.xlabel("Drinks per Week")
plt.ylabel("Drunk Driving Indicator")
plt.title("Alcohol Consumption vs Drunk Driving Risk")
plt.show()
```
